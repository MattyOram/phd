# Function to determine position of the tekscan sensor

```python
from pathlib import Path
import numpy as np
import pyvista as pv

from phd_helpers.AbaqusPostprocessing import inp2pv
```

```python
sub = '14548R'
job = '35T-neutral-00'

inp_path = Path(f'../../../Computational/InpPipeline/outputs/initialFEAstuff/accuracy/study1_35T4d5/inpFiles/{sub}/inp/{job}/{job}.inp')

meshes = inp2pv(inp_path)
tpm_mesh = meshes['tpm']
mc1_mesh = meshes['mc1']
```

```python
mc1_mesh.plot()
```

```python
def get_sensor_loc(mc1_mesh, guide_wall_z=10, sensor_offset_z=-1, sensor_size=13.97):
    """
    mc1_mesh: should be aligned with x-axis with cartilage toward negative end\n
    guide_wall_z: z offset of guide inner wall (was 10(mm) for skinny ledge and -6.9(mm) for big ledge)\n
    sensor_offset_z: z offset of sensor from guide_wall (~ -1 mm) -ive if guide_wall_z is +ive and vice verse\n
    \n
    returns: sensor centre coord (normal is (1, 0, 0))
    """

    sign = np.sign(sensor_offset_z)
    x = mc1_mesh.points[:, 0].min()
    z = guide_wall_z + (sign*(sensor_size/2)) + sensor_offset_z
    return np.array([x, 0, z])

```

```python
loc1 = get_sensor_loc(mc1_mesh, guide_wall_z=-6.9, sensor_offset_z=1, sensor_size=14)
loc2 = get_sensor_loc(mc1_mesh, guide_wall_z=10, sensor_offset_z=-1, sensor_size=14)
print('  Big loc:', loc1)
print('Small loc:', loc2)
```

```python
pl = pv.Plotter()
pl.add_mesh(mc1_mesh, color='white')
pl.add_points(loc1, color='magenta', point_size=20)
pl.add_points(loc2, color='cyan', point_size=20)
pl.show()
```
