```python
from pathlib import Path
import pyvista as pv
import numpy as np
from pathlib import Path


from phd_helpers.MeshQuality import check_mesh_quality, mesh_quality_summary
```

```python
def extract_mesh_regions(mesh):
    """mesh with bone(1) cartilage(2) volumes and bone(-1) cartilage(-2) interface(-3) surfaces"""
    shell_mesh = mesh.extract_cells_by_type(5)
    return {
        'mesh': mesh,

        'shell': shell_mesh.extract_surface(algorithm=None),
        'bone_shell': shell_mesh.extract_cells(np.where(shell_mesh['region_id']==-2)[0], invert=True).extract_surface(algorithm=None),
        'cartilage_shell': shell_mesh.extract_cells(np.where(shell_mesh['region_id']==-1)[0], invert=True).extract_surface(algorithm=None),

        'bone_surf': shell_mesh.extract_cells(np.where(shell_mesh['region_id']==-1)[0]).extract_surface(algorithm=None),
        'cartilage_surf': shell_mesh.extract_cells(np.where(shell_mesh['region_id']==-2)[0]).extract_surface(algorithm=None),
        'interface_surf': shell_mesh.extract_cells(np.where(shell_mesh['region_id']==-3)[0]).extract_surface(algorithm=None),

        'bone_vol': mesh.extract_cells(np.where(mesh['region_id']==1)[0]),
        'cartilage_vol': mesh.extract_cells(np.where(mesh['region_id']==2)[0])
    }
```

```python
# interference after cartilage generation

import os
import json
from pathlib import Path

interference_path = Path('testing3/reports/interference.jsonl')
bad_subs = []
with open(interference_path, 'r') as f:
    for line in f:
        bad_subs.append(json.loads(line)['subject'])

all_subs = [x.strip('.csv') for x in os.listdir(interference_path.parent.parent / 'meshes')]
good_subs = [x for x in all_subs if x not in bad_subs]

print('bad ', len(bad_subs), bad_subs)
print('good', len(good_subs), good_subs)
```

```python
# original mesh interference check - before any mesh processing

from phd_helpers.helpers2 import bone_pair_pose_intersection

bone_pairs = ['tpm-mc1']
poses = ['adduction','abduction','flexion','extension','pinch','grasp','jar','neutral']
safe_subs = bone_pair_pose_intersection(bone_pairs, poses,'../checkInterference2').to_list()

no_safe_subs = [x for x in all_subs if x not in safe_subs]
print('bad ', len(no_safe_subs), no_safe_subs)
print('good', len(safe_subs), safe_subs)
```

```python
shared = np.intersect1d(bad_subs, no_safe_subs)
diff = np.setxor1d(bad_subs, no_safe_subs)
print('Shared', len(shared), shared)
print('Diff  ', len(diff), diff)
```
