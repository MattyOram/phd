# Final interference subjects for TMCJ-Contact  
## With justified 2Dmesh -> cartilage parameters 

```python
import pandas as pd
import numpy as np

from phd_helpers.paths import bone_pair_pose_intersection
```

### Load the interference check for tpm and mc1 for final 2Dmesh params 
 - From: '../BonePrep/Interference/Interference-box.ipynb'

```python
df = pd.read_csv('../BonePrep/Interference/Interference-SmoothRemesh/tpm-mc1.csv', index_col=0)
df
```

#### All subjects

```python
subs_all = np.array([x.split('-')[0] + x.split('-')[1] for x in df.columns])
print(len(subs_all))
print(subs_all)
```

#### Subjects without any bone on bone interference

```python
poses = ['adduction','abduction','flexion','extension','pinch','grasp',
                            'jar','neutral']

mesh2d = np.array(
    [x.split('-')[0]+x.split('-')[1] for x in df.columns[~df[np.isin(df.index, poses)].any()].sort_values().to_list()]
    )

print(len(mesh2d))
print(mesh2d)
```

#### Subjects that fail the cartilage interference check

```python
# path to test run of final params for 
inter_path = '../../../MeshPipeline/outputs/testing/cartilage_update/reports/interference.jsonl'

inter_df = pd.read_json(inter_path, lines=True)
inter_tpm = inter_df.subject[inter_df['bones']=='tpm-mc1'].unique().to_numpy()
inter_mc1 = inter_df.subject[inter_df['bones']=='mc1-tpm'].unique().to_numpy()

ok_tpm = subs_all[~np.isin(subs_all, inter_tpm)]

print('\ntpm')
print(len(inter_tpm))
print(inter_tpm)
print('\nmc1')
print(len(inter_mc1))
print(inter_mc1)

print('\nAll subjects can generate cartilage for both bones or neither bone:', len(np.setxor1d(inter_tpm, inter_mc1))==0)
print('All subjects that pass bone interference check, pass cartilage interference check, and vice versa:', len(np.setxor1d(mesh2d, ok_tpm))==0)
```
