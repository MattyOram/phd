#### Convert the .iv files found in:
    - data/BrownUniCarpalDataset/Database/**/*IV/*
#### to .stl files found in
    - data/BrownUniCarpalDataset/Database/**/*STL/*

```python
import numpy as np
import pyvista as pv
from pathlib import Path
```

```python
#-------- iv2stl --------------------------------------
def  get_vertices_faces(bone_iv_path): 

    vertices = []
    faces = []

    get_vertices = False
    get_faces = False
    with open(bone_iv_path, 'r') as f:
        for line in f:

            if ']' in line:
                get_vertices = False
                get_faces = False


            if get_vertices:
                vertices.append([float(x) for x in line.strip().strip(',').split(' ')])

            if get_faces:
                conns = [int(x.strip()) for x in line.strip().strip(',').split(',')][:-1] #remove -1 col
                nconns = [len(conns)]
                nconns.extend(conns)  #append n-vertices to first col
                faces.append(nconns)


            if 'point [' in line:
                get_vertices = True

            if 'coordIndex [' in line:
                get_faces = True

    return np.array(vertices), np.array(faces)


def  save_stl(vertices, faces, bone_stl_path): 
    mesh = pv.PolyData(vertices, faces)
    mesh.save(bone_stl_path)

def iv2stl(bone_iv_path, bone_stl_path):
    vertices, faces = get_vertices_faces(bone_iv_path)
    save_stl(vertices, faces, bone_stl_path)
#-------- iv2stl --------------------------------------
```

```python
db_path = Path("../../data/BrownUniCarpalDataset/Database")
for path in db_path.glob("**/*.iv"):
    iv_dir = path.parent
    side = iv_dir.name[:-2]
    stl_dir = iv_dir.parent / f"{side}STL"
    stl_dir.mkdir(exist_ok=True)
    stl_path = stl_dir / path.name.replace('.iv', '.stl')
    iv2stl(path, stl_path)
```
